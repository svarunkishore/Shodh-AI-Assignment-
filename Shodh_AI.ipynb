```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
import gradio as gr


class LogicGateNN(nn.Module):
    def __init__(self):
        super(LogicGateNN, self).__init__()
        self.fc = nn.Linear(2, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.fc(x)
        x = self.sigmoid(x)
        return x


def prepare_data_gate(gate):
    if gate == "AND":
        inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)
        targets = np.array([[0], [0], [0], [1]], dtype=np.float32)
    elif gate == "OR":
        inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)
        targets = np.array([[0], [1], [1], [1]], dtype=np.float32)
    elif gate == "XOR":
        inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)
        targets = np.array([[0], [1], [1], [0]], dtype=np.float32)
    else:
        raise ValueError("Invalid gate type. Supported gates are 'AND', 'OR', and 'XOR'.")

    return torch.tensor(inputs), torch.tensor(targets)


def train_gate_model(model, inputs, targets):
    criterion = nn.BCELoss()
    optimizer = optim.SGD(model.parameters(), lr=0.1)

    epochs = 10000
    for epoch in range(epochs):
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        loss.backward()
        optimizer.step()

        if (epoch + 1) % 1000 == 0:
            print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')


def plot_decision_boundary(model, data, labels, title):
    x_min, x_max = data[:, 0].min() - 0.1, data[:, 0].max() + 0.1
    y_min, y_max = data[:, 1].min() - 0.1, data[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))

    grid = torch.tensor(np.c_[xx.ravel(), yy.ravel()], dtype=torch.float32)
    with torch.no_grad():
        probs = model(grid).reshape(xx.shape)

    plt.contourf(xx, yy, probs, alpha=0.8, cmap=plt.cm.Spectral)
    plt.scatter(data[:, 0], data[:, 1], c=labels.ravel(), edgecolor='k', s=100, cmap=plt.cm.Spectral)
    plt.xlabel('Input 1')
    plt.ylabel('Input 2')
    plt.title(title)
    plt.show()


def predict_logic_gate(input1, input2, gate):
    inputs = torch.tensor([[input1, input2]], dtype=torch.float32)
    with torch.no_grad():
        output = model(inputs)
    return "1" if output.item() > 0.5 else "0"


def create_gradio_interface(gate):
    inputs, targets = prepare_data_gate(gate)
    global model
    model = LogicGateNN()
    train_gate_model(model, inputs, targets)
    plot_decision_boundary(model, inputs.numpy(), targets.numpy(), f"{gate} Gate Decision Boundary")

    iface = gr.Interface(
        fn=lambda input1, input2: predict_logic_gate(input1, input2, gate),
        inputs=[gr.Radio([0, 1], label="Input 1"), gr.Radio([0, 1], label="Input 2")],
        outputs="text",
        title=f"{gate} Gate Neural Network",
        description=f"Interactive interface to predict the output of a {gate} gate using a trained neural network."
    )
    return iface


and_iface = create_gradio_interface("AND")
or_iface = create_gradio_interface("OR")
xor_iface = create_gradio_interface("XOR")

and_iface.launch()
or_iface.launch()
xor_iface.launch()
```
